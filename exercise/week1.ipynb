## 0. Import Packages

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas
```

## 1. Circle

```python
sns.set_style("whitegrid")

fig, ax = plt.subplots()

theta = np.linspace(0, 2*np.pi, 10000)
radius = 1
x = radius * np.cos(theta)
y = radius * np.sin(theta)

sns.scatterplot(x=x, y=y, ax=ax, color='green', s=1)

ax.set_aspect('equal')
```

## 2. Four-Leaf Clover

```python
def clover_leaf(t, a, b):
    x = a * np.cos(t) * np.sin(2*t)
    y = b * np.sin(t) * np.sin(2*t)
    return x, y

t = np.linspace(0, 2*np.pi, 1000)
leaf1_x, leaf1_y = clover_leaf(t, 5, 5)
leaf2_x, leaf2_y = clover_leaf(t + np.pi/2, 5, 5)
leaf3_x, leaf3_y = clover_leaf(t + np.pi, 5, 5)
leaf4_x, leaf4_y = clover_leaf(t + 3*np.pi/2, 5, 5)

plt.plot(leaf1_x, leaf1_y, color='green', linewidth=2)
plt.plot(leaf2_x, leaf2_y, color='green', linewidth=2)
plt.plot(leaf3_x, leaf3_y, color='green', linewidth=2)
plt.plot(leaf4_x, leaf4_y, color='green', linewidth=2)

plt.title('Four-Leaf Clover')
plt.axis('equal')
plt.axis('off')

# Show the plot
plt.show()
```

## Draw Hankel Contour

```python
epsilon = 2 # Radius of the small semicircle
R = 10   # Maximum range of integration (approximates infinity)

# 1. Integration path along the positive real axis (from epsilon to R)
x1 = np.linspace(epsilon, R, 200)
y1 = np.zeros_like(x1)

# 2. Upper semicircular arc (counterclockwise)
theta_top = np.linspace(np.pi, 0, 100)  # From π to 0
x2 = epsilon * np.cos(theta_top)
y2 = epsilon * np.sin(theta_top)

# 3. Integration path along the lower real axis (from R back to epsilon)
x3 = np.linspace(R, epsilon, 200)
y3 = np.zeros_like(x3) - 0.1  # Slight downward shift for better visibility

# 4. Lower semicircular arc (counterclockwise) returning to the real axis
theta_bottom = np.linspace(0, -np.pi, 100)  # From 0 to -π
x4 = epsilon * np.cos(theta_bottom)
y4 = epsilon * np.sin(theta_bottom)
```

```python
# Create figure and axis
fig, ax = plt.subplots(figsize=(7, 5))

# Plot the Hankel contour
ax.plot(x1, y1, 'b', linewidth=2)  # Positive real axis
ax.plot(x2, y2, 'b', linewidth=2)  # Upper semicircular arc
ax.plot(x3, y3, 'b', linewidth=2)  # Lower real axis
ax.plot(x4, y4, 'b', linewidth=2)  # **Lower semicircular arc (added)**

# Add arrows to indicate the direction of integration
ax.annotate("", xy=(3, 0), xytext=(5, 0), arrowprops=dict(arrowstyle="->", color="b"))
ax.annotate("", xy=(7, -0.1), xytext=(5, -0.1), arrowprops=dict(arrowstyle="->", color="b"))
ax.annotate("", xy=(epsilon * np.cos(np.pi/4), epsilon * np.sin(np.pi/4)),
            xytext=(epsilon * np.cos(np.pi/3), epsilon * np.sin(np.pi/3)),
            arrowprops=dict(arrowstyle="->", color="b"))
ax.annotate("", xy=(epsilon * np.cos(-np.pi/4), epsilon * np.sin(-np.pi/4)),
            xytext=(epsilon * np.cos(-np.pi/3), epsilon * np.sin(-np.pi/3)),
            arrowprops=dict(arrowstyle="->", color="b"))

# Mark key points
ax.scatter([epsilon, R], [0, 0], color='r', label="Key Points")  # Important points (endpoints of arcs)

# Add labels
ax.text(epsilon, -0.3, r'$\epsilon$', fontsize=12, color='r')
ax.text(R, -0.3, r'$+\infty$', fontsize=12, color='r')
ax.text(0.5, 1.2, r'$\gamma(\epsilon)$', fontsize=12, color='b')

# Draw coordinate axes
ax.axhline(0, color='gray', linestyle='--', linewidth=0.5)
ax.axvline(0, color='gray', linestyle='--', linewidth=0.5)

# Set title and labels
ax.set_title("Hankel Contour", fontsize=14)
ax.set_xlabel("Re(s)")
ax.set_ylabel("Im(s)")

# Adjust display range
ax.set_xlim(-2, R + 2)
ax.set_ylim(-2, 2)
ax.legend()
plt.axis('equal')
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```
