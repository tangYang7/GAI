```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
sns.set_style("whitegrid")

fig, ax = plt.subplots()

theta = np.linspace(0, 2*np.pi, 10000)
radius = 1
x = radius * np.cos(theta)
y = radius * np.sin(theta)

sns.scatterplot(x=x, y=y, ax=ax, color='green', s=1)

ax.set_aspect('equal')
```

```python
def clover_leaf(t, a, b):
    x = a * np.cos(t) * np.sin(2*t)
    y = b * np.sin(t) * np.sin(2*t)
    return x, y

t = np.linspace(0, 2*np.pi, 1000)
leaf1_x, leaf1_y = clover_leaf(t, 5, 5)
leaf2_x, leaf2_y = clover_leaf(t + np.pi/2, 5, 5)
leaf3_x, leaf3_y = clover_leaf(t + np.pi, 5, 5)
leaf4_x, leaf4_y = clover_leaf(t + 3*np.pi/2, 5, 5)

plt.plot(leaf1_x, leaf1_y, color='green', linewidth=2)
plt.plot(leaf2_x, leaf2_y, color='green', linewidth=2)
plt.plot(leaf3_x, leaf3_y, color='green', linewidth=2)
plt.plot(leaf4_x, leaf4_y, color='green', linewidth=2)

plt.title('Four-Leaf Clover')
plt.axis('equal')
plt.axis('off')

plt.show()
```

```python
def clover_leaf(t, a, b):
    x = a * np.cos(t) * np.sin(2*t)
    y = b * np.sin(t) * np.sin(2*t)
    return x, y

t = np.linspace(0, 2*np.pi, 1000)
leaf1_x, leaf1_y = clover_leaf(t, 5, 5)
leaf2_x, leaf2_y = clover_leaf(t + np.pi/2, 5, 5)
leaf3_x, leaf3_y = clover_leaf(t + np.pi, 5, 5)
leaf4_x, leaf4_y = clover_leaf(t + 3*np.pi/2, 5, 5)

sns.set_style("white")  # Set a clean background
fig, ax = plt.subplots()

sns.lineplot(x=leaf1_x, y=leaf1_y, ax=ax, color='green', linewidth=2)
sns.lineplot(x=leaf2_x, y=leaf2_y, ax=ax, color='green', linewidth=2)
sns.lineplot(x=leaf3_x, y=leaf3_y, ax=ax, color='green', linewidth=2)
sns.lineplot(x=leaf4_x, y=leaf4_y, ax=ax, color='green', linewidth=2)

ax.set_title('Four-Leaf Clover')
ax.set_aspect('equal')
ax.axis('off')
```
